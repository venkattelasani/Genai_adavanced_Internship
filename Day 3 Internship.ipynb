```python
#Write a Python program to get a sorted list in increasing order of last element in each tuple in a given list using loop.
#Input:- [(5,4),(9,1),(2,3),(5,9),(7,6),(5,5)]
#output:- [(9,1),(2,3),(5,4),(5,5),(7,6),(5,9)]

# Using Bubble sort techique 

lst = [(5,4),(9,1),(2,3),(5,9),(7,6),(5,5)]
for i in range(len(lst)): # range(0-5)
    for j in range(len(lst)-i-1): # range(5)
        if lst[j][-1] > lst[j+1][-1]:
            lst[j], lst[j+1] = lst[j+1], lst[j]

print(lst)
```

```python
#Write a Python program to create a 4X4 2D matrix with below elements using loop and list comprehension both.
#Output:- [[0,0,0,0],[0,1,2,3],[0,2,4,6],[0,3,6,9]]

# 1-D array ---> 1 list ---> []
# 2-D array ---> 2 1-D list in a list ---> [[], []]
# 3-D array ----> 2 2-D list in a list - --> [[[], []],  [[], []]]

a = [] # outer list
b = [] # inner list 

for i in range(4): # rows 
    for j in range(4): # columns
        b.append(i*j)
    a.append(b)
    b = []

print(a)
```

```python
#Q13. Write a function that converts a sparse vector into a dictionary as described above.
# In - convertVector([1, 0, 0, 2, 0, 0, 0, 3, 0, 0, 0, 0, 4])
# Out - {0: 1, 3: 2, 7: 3, 12: 4}
# In - convertVector([1, 0, 1 , 0, 2, 0, 1, 0, 0, 1, 0])
# Out - {0: 1, 2: 1, 4: 2, 6: 1, 9: 1}
# In - convertVector([0, 0, 0, 0, 0])
# Out - {}

def convert_vector(lst):
    d = {}
    count = 0 # taking care of the index
    for i in lst:
        if i != 0:
            d[count] = i
        count+= 1
    return d 
```

```python
convert_vector([1, 0, 0, 2, 0, 0, 0, 3, 0, 0, 0, 0, 4])
```
