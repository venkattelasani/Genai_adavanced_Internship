```python
def mimic_capitalize(text):
    if not text:
        return text

    result = ""

    first = text[0]
    if 'a' <= first <= 'z':
        first = chr(ord(first) - 32)
    result += first
    
    for ch in text[1:]:
        if 'A' <= ch <= 'Z':
            ch = chr(ord(ch) + 32)
        result += ch

    return result
```

```python
print(mimic_capitalize("pYtHoN CODE"))      
print(mimic_capitalize("hello WORLD"))       
print(mimic_capitalize("@JAVA PROGRAM"))     
```

```python
def mimic_title(text):
    result = ""
    new_word = True

    for ch in text:
        if ch == " ":
            new_word = True
            result += ch
        else:
            if new_word:
                
                if 'a' <= ch <= 'z':
                    ch = chr(ord(ch) - 32)
                new_word = False
            else:
                
                if 'A' <= ch <= 'Z':
                    ch = chr(ord(ch) + 32)
            result += ch
    return result
```

```python
print(mimic_title("hello world from python"))
print(mimic_title("mACHINE lEARNING is FUN"))
print(mimic_title("data-SCIENCE  with python"))
```

```python
def mimic_swapcase(text):
    result = ""
    for ch in text:
        if 'a' <= ch <= 'z':       
            result += chr(ord(ch) - 32)
        elif 'A' <= ch <= 'Z':      
            result += chr(ord(ch) + 32)
        else:
            result += ch            
    return result
```

```python
print(mimic_title("hello world from python"))
print(mimic_title("mACHINE lEARNING is FUN"))
print(mimic_title("data-SCIENCE  with python"))
```

```python
def mimic_isupper(text):
    has_alpha = False
    for ch in text:
        if 'a' <= ch <= 'z':
            return False
        if 'A' <= ch <= 'Z':
            has_alpha = True
    return has_alpha
```

```python
print(mimic_isupper("HELLO"))      
print(mimic_isupper("Hello"))     
print(mimic_isupper("123!"))       
```

```python
def mimic_isalpha(text):
    if not text:
        return False
    for ch in text:
        if not ('a' <= ch <= 'z' or 'A' <= ch <= 'Z'):
            return False
    return True
```

```python
print(mimic_isalpha("Python"))    
print(mimic_isalpha("abc123"))     
print(mimic_isalpha("Data_Science")) 
```

```python
def mimic_lstrip(text):
    i = 0
    while i < len(text) and text[i] == " ":
        i += 1
    return text[i:]
```

```python
print(mimic_lstrip("    python"))
```

```python
def mimic_rstrip(text):
    i = len(text) - 1
    while i >= 0 and text[i] == " ":
        i -= 1
    return text[:i+1]
```

```python
print(mimic_rstrip("python    "))
```

```python
def mimic_isdigit(text):
    if not text:
        return False
    for ch in text:
        if not ('0' <= ch <= '9'):
            return False
    return True
```

```python
print(mimic_isdigit("12345"))  
print(mimic_isdigit("12a45")) 
```
